## Оборудование и реактивы

В качестве радикала используется ABTS-радикал (2,2’-azino-bis(3-ethylbenzothiazoline-sulfonicacid), пероксидисульфат калия, Na2HPO4, NaH2PO4 для приготовления буферного раствора (0.1М, pH=7.4: 0.0754М и 0.0246М, соответственно). В качестве стандартного антиоксиданта используется Trolox ((±)-6-Hydroxy-2,5,7,8-tetramethylchromane-2-carboxylic acid). Для приготовления рабочих растворов используют дегазированную дистиллированную воду. Для этого либо пропускают инертный газ через неё, либо выдерживают при низком давлении в течение 30 минут.

Эппиндорфы, штатив, автоматические пипетки, стаканчики для приготовления проб и измерений. Спектрофотометр, оперирующий в области 700-800 нм.


## Приготовление рабочего раствора ABTS-радикала

1.	Ввешиваем точную навеску АБТС – 11 мг в эпиндорфе. Добавляем 900 мкл дистиллированной воды и перемешиваем на вортоксе.
2.	Взвешиваем 20 мг пероксидисульфата калия в эпиндорфе. Добавляем 1000 мкл дистиллированной воды, перемешиваем на вортоксе
3.	Далее 100 мкл раствора пероксидисульфата калия добавляем в раствор АБТС и перемешиваем
4.	Через 12 часов или более раствор разбавляем фосфатным буфером (0.1 М, рН 7.4) до поглощения 0.7 при длине волны 734 нм.

## Приготовление стандартного раствора Trolox

1.	Тщательно взвешиваем 2.5 мг тролокса в эпиндорфе. Добавляем 500 мкл этилового спирта и перемешиваем на вортоксе.
2.	Количественно переносим в 100 мл колбу, разбавляем водой.
3.	Получаем рабочий раствор с концентрацией 100 мкмоль/литр

## Приготовление растворов гуминовых веществ

1.	Взвешиваем 10 мг препарата и переносим в 100 мл мерную колбу для получение рабочей концентрации в 100 мг/л
2.	Если препарат плохо растворим, добавляем в начале несколько капель концентрированной щелочи (NaOH).
3.	Доводим до метки с помощью дегазированной дистиллированной воды

## Измерение

1.	Для измерения в планшет раскладываем эпиндорфы, по 3 на каждый образец, стандарт и ещё три для контрольных измерений.
2.	Образец раскапываем в эпиндорфы: 200 мкл, 140 мкл, 80 мкл. Затем туда же добавляем воду, доведя общий объем до 200 мкл (0, 60 и 120 мкл, соответственно).
3.	Стандарт раскапываем в эпиндорфы: 200, 140, 100, 60, 40 мкл. Затем туда же добавляем воду, доведя общий объем до 200 мкл (0, 60, 100, 140, 160 мкл, соответственно)
4.	В контроль добавляем по 200 мкл воды.
5.	Добавляем 1800 мкл рабочего раствора АБТС в каждый эпиндорф, перемешиваем на вортоксе.
6.	Спустя 40 минут после добавления АБТС регистрируем спектры поглощения 400 – 800 нм.

## Расчёт антиоксидантной емкости

1.	В каждом спектре определяем оптическое поглощение при 734 нм.
2.	Получаем разность между поглощением контроля (воды) и поглощением образца
3.	Строим градуировочную зависимость для стандарта в координатах: мкмоль/л — разница оптического поглащения.
4.	Определяем TEAC (Trolox equivalent antioxidant capacity) для всех образцов согласно полученной градуировке в единицах мкмоль/мг

# Обработка результатов определения антиоксидантной активности

Загружаем данные со спектрометра - сырой файл. 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data = pd.read_csv('data/exampl2.csv')
data.head()
```

Удаляем все нечетные столбцы где дублируется значение волны поглащения, оставим только первый.

```python
data = data.drop([data.columns[x] for x in range(2, len(data.columns), 2)], axis=1)
data.head()
```

Переименовываем колонки согласно последовательности как мы измеряли, первое значение оставим длине волны поглащения

```python
column = ['wave',
          'b1', 'b2', 'b3',
          'j1', 'j2', 'j3',
          't1', 't2', 't3',
          'L1', 'L2', 'L3',
          'blank','blank','blank',
          's1','s2','s3','s4','s5'        
        ]

data.set_axis(column, axis=1, inplace=True)
```

Сохраняем только строку со значениями поглащения при 734 нм, на основании которых будем считать результаты

```python
data = data.drop([0])
data['wave'] = pd.to_numeric(data['wave']).astype('int')
data = data.loc[data['wave'] == 734]
data = data.reset_index()
data = pd.DataFrame({'sample':data.columns[2:], 'absorbance':data.iloc[0].tolist()[2:]})
data['absorbance'] = pd.to_numeric(data['absorbance'])
data
```

Высчитываем среднее поглощение холостого раствора АБТС-радикала - blank

```python
blank_mean = data[data['sample'] == 'blank']['absorbance'].mean()
```

находим разницу между поглощением бланка и образцами
удаляем значения бланка

```python
data['delta'] = blank_mean - data['absorbance']
data = data.loc[data['sample'] != 'blank']
data
```

Добавляем столбец концентраций, получившихся в соответствующих эпиндорфах после добавления образцов и воды, но до добавления раствора радикала. Для образцов - в мг/л, для тролокса в мкмоль/л

```python
volume = [50, 35, 20,
          50, 35, 20,
          50, 35, 20,
          50, 35, 20,
          400, 280, 200, 120, 80
         ]

data.insert(3, 'conc', volume)
data
```

Выбираем данные для каллибровки, добавим ноль

```python
cal = data.loc[18:19, ('delta', 'conc')]
cal = cal.append(pd.Series({'delta': 0, 'conc': 0}), ignore_index=True)
```

Построим каллибровочный график и аппроксимируем его прямой

```python
plt.plot(cal['conc'], cal['delta'], 'o', color='black')
params = np.polyfit(cal['conc'], cal['delta'], 1)
xp = np.linspace(0, 120, 10)
yp = np.polyval(params, xp)
plt.plot(xp, yp, 'g')
```

Каллибровка хоть и по двум реальныйм точкам, но нормальная. Посмотрим на коэффциенты аппроксимирующего уровнения y = a*x + b

```python
a = params[0]
b = params[1]
a, b
```

Убирем из данных стандарт. Рассчитываем антиоксидантную емкость в единицах мкмоль тролокс-эквивалента на мг образца.
Для этого нужно поделить разницу поглащений на концентрацию и на коээфциента а из уравнения

```python
data = data.loc[0:11, :]
data = data.reset_index(drop = True)
data['TEAC'] = data['delta'] / (data['conc'] * a)
data
```

Видно, что образец L зашкалил, результаты по нему не получились.

Создаим новый столбец с названием, без последнего символа, чтобы далее сгруппировать данные по образцу

```python
data['name'] = data['sample'].apply(lambda name: name[:-1])
```

Сведём данные в одну таблицу, посчитаем среднее и доверительный интервал.

```python
data_mean = data.loc[:, ['name', 'TEAC']].groupby('name').mean()
data_std = data.loc[:, ['name', 'TEAC']].groupby('name').std()
data_std.columns = ['std']
result = data_mean.merge(data_std, on='name')
result
```
